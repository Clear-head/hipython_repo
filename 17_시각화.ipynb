```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
# 1번량 그래프: 히스토그램
fig = plt.figure(figsize=(12, 5))

axe1 = fig.add_subplot(1, 2, 1)
axe2 = fig.add_subplot(1, 2, 2)


axe1.hist(
    x=tips_df['total_bill'],
    bins=20,
    label="total_bill"
    )

axe2.hist(
    x=tips_df['tip'],
    bins=20,
    label="tip"
    )

plt.tight_layout()
```

```python
# box plot


fig = plt.figure(figsize=(12, 5))

axe1 = fig.add_subplot(1, 2, 1)
axe2 = fig.add_subplot(1, 2, 2)


axe1.boxplot(x=tips_df['total_bill'])

axe2.boxplot(x=tips_df['tip'])

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(12, 5))

axes = fig.add_subplot(1, 1, 1)

axes.boxplot(
    x = [
        tips_df[tips_df['sex']=='Female']['total_bill'],
        tips_df[tips_df['sex']=='Male']['total_bill']
    ]
)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(12, 5))

axes = fig.add_subplot(1, 1, 1)
tmp = set(tips_df['day'])

axes.boxplot(
    x = [
        tips_df[tips_df['day']==i]['total_bill'] for i in tmp
    ]
)

plt.tight_layout()
```

```python
# scatter, 산점도
fig = plt.figure(figsize=(12, 5))
axes = fig.add_subplot(1, 2, 1)
axes.scatter(
    x= tips_df['total_bill'], y= tips_df['tip']
)
plt.tight_layout()
```

```python
# 다변량 표현
# 성별에 따라 크기, 색상
# 값의 크기에 따라 - 점의 크기
colors = {"Female":"Red", "Male":"Blue"}
fig = plt.figure(figsize=(12, 5))
axes1 = fig.add_subplot(1, 1, 1)
axes2 = fig.add_subplot(2, 1, 1)
axes3 = fig.add_subplot(3, 1, 1)
axes4 = fig.add_subplot(4, 1, 1)

axes1.scatter(
    x = tips_df[tips_df['day'] == 'Thur']['total_bill'], 
    y = tips_df[tips_df['day'] == 'Thur']['tip'],
    s=tips_df[tips_df['day'] == 'Thur']['tip']**2 * 10,
    c=tips_df[tips_df['day'] == 'Thur']['sex'].map(colors)
)
axes2.scatter(
    x = tips_df[tips_df['day'] == 'Sat']['total_bill'], 
    y = tips_df[tips_df['day'] == 'Sat']['tip'],
    s=tips_df[tips_df['day'] == 'Sat']['tip']**2 * 10,
    c=tips_df[tips_df['day'] == 'Sat']['sex'].map(colors)
)
axes3.scatter(
    x = tips_df[tips_df['day'] == 'Fri']['total_bill'], 
    y = tips_df[tips_df['day'] == 'Fri']['tip'],
    s=tips_df[tips_df['day'] == 'Fri']['tip']**2 * 10,
    c=tips_df[tips_df['day'] == 'Fri']['sex'].map(colors)
)
axes4.scatter(
    x = tips_df[tips_df['day'] == 'Sun']['total_bill'], 
    y = tips_df[tips_df['day'] == 'Sun']['tip'],
    s=tips_df[tips_df['day'] == 'Sun']['tip']**2 * 10,
    c=tips_df[tips_df['day'] == 'Sun']['sex'].map(colors)
)
# plt.tight_layout()
```

```python
# 파이차트: 비율
# 성별 비율 비교
# 빈도수
fig = plt.figure(figsize=(12, 5))
axes = fig.add_subplot(1, 2, 1)
axes.pie(
    tips_df['sex'].value_counts(),
    colors=["Red", "Blue"],
    labels=tips_df['sex'].value_counts().index,
    autopct="%1.1f%%",
    startangle=90
)
plt.tight_layout()
```

```python
# 요일별

fig = plt.figure(figsize=(12, 5))
axes = fig.add_subplot(1, 2, 1)
axes.pie(
    tips_df['day'].value_counts(),
    labels=tips_df['day'].value_counts().index,
    autopct="%1.1f%%",
    startangle=90
)
plt.tight_layout()
```

```python
# 흡연자 비율

fig = plt.figure(figsize=(12, 5))
axes = fig.add_subplot(1, 2, 1)
axes.pie(
    tips_df['smoker'].value_counts(),
    labels=tips_df['smoker'].value_counts().index,
    autopct="%1.1f%%",
    startangle=90
)
plt.tight_layout()
```

```python
# 점심저녁 비율

fig = plt.figure(figsize=(12, 5))
axes = fig.add_subplot(1, 2, 1)
axes.pie(
    tips_df['time'].value_counts(),
    labels=tips_df['time'].value_counts().index,
    autopct="%1.1f%%",
    startangle=90
)
plt.tight_layout()
```
