### 월급 계산

```python
# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python
def ages(d):
    if d < 30:
        return "20s"
    elif d < 40:
        return "30s"
    elif d < 50:
        return "40s"
    elif d < 60:
        return "50s"
    elif d < 70:
        return "60s"
    else:
        return "too old"
```

```python
df["Category"] = df["Age"].apply(ages)
```

```python
def my_func(a, a_mean):
    if a < a_mean:
        return "low"
    elif a == a_mean:
        return "middle"
    else:
        return "high"
```

```python
df["tmp"] = df["Salary"].apply(my_func, a_mean=df["Salary"].mean())
```
