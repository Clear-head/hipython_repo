## 교차 검증

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5)
iris_clf = dtc(random_state=156)

iris.data, iris.target
```

```python
for train_idx, test_idx in kfold.split(iris.data):
    x_train, x_test = iris.data[train_idx], iris.data[test_idx]
    y_train, y_test = iris.target[train_idx], iris.target[test_idx]

    iris_clf.fit(x_train, y_train)
    pred = iris_clf.predict(x_test)
    accuracy = accuracy_score(y_test, pred)
    print(f"Accuracy: {accuracy:.4f}")
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(iris.data, iris.target):
  y_train = iris.data[train_index]
  y_test =  iris.target[test_index]
  print(f'train 분포: {y_train}')
  print(f'test 분포: {y_test}')
```

# GridSearchCV

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=121)
```

```python
from sklearn.model_selection import GridSearchCV

dt_clf2 = dtc(random_state=121)

params = {'max_depth' : [1, 2, 3, 4], 'min_samples_split' : [2, 3, 4]}
grid_dtree = GridSearchCV(dt_clf2, param_grid=params, cv=5, refit=True)

grid_dtree.fit(x_train, y_train)
```

```python
pd.DataFrame(grid_dtree.cv_results_)
```

```python
grid_dtree.best_params_
```

```python
pred = grid_dtree.predict(x_test)
accuracy_score(y_test, pred)
```

## 피처 스케일링

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data = iris.data, columns=iris.feature_names)
iris_df
```

```python
iris_df.var()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
iris_scaled.mean()
```

```python
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(iris.data, 
                                                    iris.target, 
                                                    test_size=0.2, 
                                                    random_state=11 )
```

```python
no_scaling_model = LogisticRegression(random_state=11)
no_scaling_model.fit(X_train, y_train)
pred_1 = no_scaling_model.predict(X_test)
acc_1 = accuracy_score(y_test, pred_1)
acc_1
```

```python
from sklearn.preprocessing import StandardScaler
scaler1 = StandardScaler()
scaler1.fit(X_train)
X_train_scaled = scaler1.transform(X_train)
X_test_scaled = scaler1.transform(X_test)
```

```python
scaling_model = LogisticRegression(random_state=11)
scaling_model.fit(X_train_scaled, y_train)
pred_2 = scaling_model.predict(X_test_scaled)
acc_2 = accuracy_score(y_test, pred_2)
acc_2, acc_1
```
