# 상관분석

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
```

```python
df = pd.read_csv("./data/datalab_stress.csv")
df
```

```python
df[["스트레스", "집값", "우울증"]].corr()
```

```python
corr1, p_value1 = pearsonr(df['스트레스'], df['집값'])

print("스트레스, 집값")
print(f"피어슨 상관계수: {corr1:.2f}")
print(f"p-값: {p_value1:.4f}")

rho, pval = spearmanr(df['스트레스'], df['집값'])

print(f"스피어만 상관계수: {rho:.3f}")
print(f"p-value: {pval:.4f}")
```

```python
corr1, p_value1 = pearsonr(df['스트레스'], df['우울증'])

print("스트레스, 우울증")
print(f"피어슨 상관계수: {corr1:.2f}")
print(f"p-값: {p_value1:.4f}")

rho, pval = spearmanr(df['스트레스'], df['우울증'])

print(f"스피어만 상관계수: {rho:.3f}")
print(f"p-value: {pval:.4f}")
```

```python
corr1, p_value1 = pearsonr(df['스트레스'], df['코스피'])

print("스트레스, 코스피")
print(f"피어슨 상관계수: {corr1:.2f}")
print(f"p-값: {p_value1:.4f}")

rho, pval = spearmanr(df['스트레스'], df['코스피'])

print(f"스피어만 상관계수: {rho:.3f}")
print(f"p-value: {pval:.4f}")
```
